The first step will be to clean and normalise the training data.

```python
import pandas as pd

#import training dataset and testing dataset
df_train = pd.read_csv('AG News Dataset/train.csv')
df_test = pd.read_csv('AG News Dataset/test.csv')

#check dataframe structures
print(df_train.head())
print(df_test.head())
```

```python
#drop all the empty columns 
df_train = df_train.drop(['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15'], axis=1)
df_test = df_test.drop(['Category','Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15'], axis=1)

#check the training data
print(df_train.head(10))
```

```python
from sklearn.preprocessing import LabelEncoder

#import label encoder and apply it to the category column
label_encoder = LabelEncoder()
df_train['Category_encoded'] = label_encoder.fit_transform(df_train['Category'])
df_train['Category_encoded'] = df_train['Category_encoded'] + 1

# Sample 5% of the data for testing - anything higher crashes my computer, explination in report
df_train = df_train.sample(frac=0.05, random_state=42)

#check training data
print(df_train.head(10))
```

```python
#check the encoding values to make sure there are the same quantity as there are categories
unique_values = df_train['Category_encoded'].unique()
print("Unique encoded values:", unique_values)

#line break for clarity
print("\n")

#Group up one of each category as an example and display it 
examples = df_train.groupby('Category_encoded').first().reset_index()
print(examples)
```

```python
#extract info from training data and check for null fields
print(df_train.info)
print(df_train.isnull().sum())
```

```python
#although there are no null fields, dropna just incase data input changes
df_train = df_train.dropna()
print(df_train.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer

#vectorise the title category so that we have numerical representations of the text
vectorizer = TfidfVectorizer()
vec_title = vectorizer.fit_transform(df_train['Title']).toarray()

#store the vecorised data in the X variable
X = pd.DataFrame(vec_title)
#store the encoded training data in the Y variable
Y = df_train['Category_encoded']

#load and apply the scaler to the X axis data and check the output
scaler = StandardScaler()
X = scaler.fit_transform(X)
print("\nNormalized features:")
print(X)
print(Y.head())
```

```python
#debugging, trying to find missing values, i added outputting the shape to find
print(f"\nShape of X: {X.shape}")
print(f"Shape of Y: {Y.shape}")
```

```python
#debugging this codeblock, I discovered program crashing kernel during test_train_split
print("debug Point 1 successful")

#load KNN calssifier method from SKLearn framework
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

#debugging
print("debug Point 2 successful")

#split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

#debugging continues
print("debug Point 3 successful")
```

```python
#create and train the KNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

```

```python
#predict on the test set from train_test_split test data comes later 
y_pred_knn = knn.predict(X_test)

```

```python
#evaluate the model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)
class_report_knn = classification_report(y_test, y_pred_knn)

#human readible output
print(f"\nKNN Accuracy: {accuracy_knn * 100:.2f}%")
print("KNN Confusion Matrix:")
print(conf_matrix_knn)
print("KNN Classification Report:")
print(class_report_knn)
```

```python
from sklearn.linear_model import LogisticRegression

#create and train the Logistic Regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
```

```python
#predict on the test set from train_test_split
y_pred_log_reg = log_reg.predict(X_test)
```

```python
#evaluate the model
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
conf_matrix_log_reg= confusion_matrix(y_test, y_pred_log_reg)
class_report_log_reg = classification_report(y_test, y_pred_log_reg)

#human readible output
print(f"\nLogistic Regression Accuracy: {accuracy_log_reg * 100:.2f}%")
print("Confusion Matrix:")
print(conf_matrix_log_reg)
print("Classification Report:")
print(class_report_log_reg)
```

```python
from sklearn.tree import DecisionTreeClassifier

#create and train the Decision Tree model
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
```

```python
#predict on the test set from train_test_split
y_pred_decision_tree = decision_tree.predict(X_test)
```

```python
#evaluate the model
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
conf_matrix_decision_tree = confusion_matrix(y_test, y_pred_decision_tree)
class_report_decision_tree = classification_report(y_test, y_pred_decision_tree)

#human readible output
print(f"\nDecision Tree Accuracy: {accuracy_decision_tree * 100:.2f}%")
print("Decision Tree Confusion Matrix:")
print(conf_matrix_decision_tree)
print("Decision Tree Classification Report:")
print(class_report_decision_tree)
```

```python
from sklearn.svm import SVC

#create and train the SVM model
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
```

```python
#predict on the test set
y_pred_svm_model = svm_model.predict(X_test)
```

```python
#evaluate the model
accuracy_svm_model = accuracy_score(y_test, y_pred_svm_model)
conf_matrix_svm_model = confusion_matrix(y_test, y_pred_svm_model)
class_report_svm_model = classification_report(y_test, y_pred_svm_model)

#human readible output
print(f"\nSVM Accuracy: {accuracy_svm_model * 100:.2f}%")
print("SVM Confusion Matrix:")
print(conf_matrix_svm_model)
print("SVM Classification Report:")
print(class_report_svm_model)
```

```python
#vectorize the 'Title' column in the test data
vec_title_test = vectorizer.transform(df_test['Title']).toarray()

#normalize the features in test data
X_test_final = scaler.transform(vec_title_test)

#predict the categories for the test data using trained model log_reg (best accuracy from HRO)
y_test_pred = log_reg.predict(X_test_final)

#convert integer predictions to category labels for a excel comparison
category_mapping = {1: "World", 2: "Sports", 3: "Business", 4: "Sci/Tech"}
df_test['Category_predicted'] = y_test_pred
df_test['Category_predicted'] = df_test['Category_predicted'].map(category_mapping)

#reorder the columns to have 'Category_predicted' first so CSV output is similar to test data
columns = ['Category_predicted'] + [col for col in df_test.columns if col != 'Category_predicted']
df_test = df_test[columns]

#output the results to a CSV file 
df_test.to_csv('test_predictions.csv', index=False)

#nice output to tell us processing is complete
print("\nPredictions on test data have been saved to 'test_predictions.csv'.")
print("You have found the last Dragonball")
```
